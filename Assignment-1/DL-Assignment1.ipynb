```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# Digit classification
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values (0–255) to range 0–1
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten 28x28 images into 1D vectors of 784
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)
```

```python
model = Sequential()

model.add(Dense(256, activation='relu', input_shape=(784,)))  #dense is like nn.linear in pytorch.
#In Python, (784) is just a number.
#(784,) is a tuple with one element.
#Keras expects the shape as a tuple.
model.add(Dropout(0.25))


# Hidden layer 2
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.25))

# Output layer (10 classes for digits 0–9)
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test accuracy:", test_accuracy)
```
